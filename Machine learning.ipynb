# Machine language

### Is a field of building algorithm to train computer from learning from the past event in order to predict for future event



### Types of machine Learning
##### 1. Supervised learning: The algorithm is trained on labeled data. which means the inputs is paired with the output.The main goal is to make prediction to learning the pattern within the input data

### Type of technique on SL
 ##### .Regression(is used to predict continous variable)
 ##### Classification (is used to predict discrete variable)

##### 2. Unsupervised Learning: The algorithm is trained and unlabelled data.  The goals is to find the hidden pattern on the data or intrinsic structure in the input data

### Type of technique on USL
##### Clustering: This is group data into cluster based on similarity 
##### Association: This discovers rules that describe larger portion of the data
##### Dimensionality Reduction: Reduce the number of random variable under consideration

#### 3.Semi- Supervised Learning:This combine small amount of label data and larger amount of unlabelled data during training

### Type of technique on SSL
###### Self training algorithms
###### co training
###### generative model

#### Renforcement : This involved an agent that learns to make decision by performing actions in an environment to maximize cummulative reward. the agent learn from the consequence of its actions, rather than from being told what to do.
