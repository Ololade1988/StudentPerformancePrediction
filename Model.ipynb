```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error
```

```python
df=pd.read_csv("train.csv")
```

# Encode all the category columns

```python
df.columns
```

```python
ordinal_cat =["famsize",'Medu','Fedu',"traveltime","health","famrel","freetime","goout","Dalc","Walc","health"]
nominal_cat =["school","sex","address","Pstatus","Mjob",'Fjob',"reason","guardian","failures","schoolsup","famsup","paid","activities","nursery","higher","internet","romantic"]
numerical =["age","absences"]
```

```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
le.fit_transform(df["famsize"])
```

```python
df_1=df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
for i in ordinal_cat:
    df_1[i]=le.fit_transform(df[i])
df_1
```

```python
df_1=pd.get_dummies(df_1,columns=nominal_cat,drop_first=True)
```

```python
# split the data into features and target
x=df_1.copy()
y=x.pop("Grade")
```

```python
x_train,x_val,y_train,y_val=train_test_split(x,y,random_state=42,test_size=0.2)
```

# First Algorithm

```python
lr=LinearRegression()
model_1=lr.fit(x_train,y_train)

predict_1= model_1.predict(x_val)
eval_1= mean_absolute_error(y_val,predict_1)
print(f"the mean absolute error by {lr} is {eval_1}")
```

```python
ls=Lasso()
model_2=ls.fit(x_train,y_train)

predict_2= model_2.predict(x_val)
eval_2= mean_absolute_error(y_val,predict_2)
print(f"the mean absolute error by {ls} is {eval_2}")
```

```python
dr=DecisionTreeRegressor()
model_3=dr.fit(x_train,y_train)

predict_3= model_3.predict(x_val)
eval_3= mean_absolute_error(y_val,predict_3)
print(f"the mean absolute error by {dr} is {eval_3}")
```

```python
sv=SVR()
model_4=sv.fit(x_train,y_train)

predict_4= model_4.predict(x_val)
eval_4= mean_absolute_error(y_val,predict_4)
print(f"the mean absolute error by {sv} is {eval_4}")
```

```python
ab=AdaBoostRegressor()
model_5=ab.fit(x_train,y_train)

predict_5= model_5.predict(x_val)
eval_5= mean_absolute_error(y_val,predict_5)
print(f"the mean absolute error by {ab} is {eval_5}")
```

```python
gb=GradientBoostingRegressor()
model_6=gb.fit(x_train,y_train)

predict_6= model_6.predict(x_val)
eval_6= mean_absolute_error(y_val,predict_6)
print(f"the mean absolute error by {gb} is {eval_6}")
```

```python
def back_elimination(y,x,significant_level=0.05):
    while True:
        model = sm.OLS(y,x).fit()
        all_pvalues = model.pvalues
        max_pvalue = all_pvalues.max()
        if max_pvalue > significant_level:
            col_to_remove = all_pvalues.idxmax()
            x = x.drop(columns=[col_to_remove])
        else:
            break
    return model
```

```python
signal_features=back_elimination(y,x)
```

```python
from sklearn.model_selection import RandomizedSearchCV

grim ={ "n_estimators":[100, 1000], "loss" :['squared_error','absolute_error'], "min_samples_leaf" :[1,2]}
model_rand = RandomizedSearchCV(estimator=gb,param_distributions=grim, cv=5, n_jobs=-1)

model_rand.fit(x_train, y_train)

best_model = model_rand.best_estimator_

model_6=best_model.fit(x_train,y_train)

predict_6= model_6.predict(x_val)
eval_6= mean_absolute_error(y_val,predict_6)
print(f"the mean absolute error by {gb} is {eval_6}")


```
